```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

%matplotlib inline
```

## Categorical scatterplots

```python
tips = sns.load_dataset("tips")
sns.catplot(
    data=tips, x="day", y="total_bill"
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill",
    jitter=False
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill",
    kind="swarm"
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="swarm"
)
```

```python
sns.catplot(
    data=tips.query("size != 3"), x="size", y="total_bill"
)
```

"order" parameter

```python
sns.catplot(
    data=tips, x="smoker", y="tip",
    order=["No", "Yes"]
)
```

Swapping the assignment of variables at axes

```python
sns.catplot(
    data=tips, x="total_bill", y="day",
    hue="time", kind="swarm"
)
```

## Comparing distributions

#### Boxplots

```python
sns.catplot(
    data=tips, x="day", y="total_bill",
    kind="box"
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="smoker",
    kind="box"
)
```

Boxenplot() -> similar to a box plot but optimized for showing more information about the shape of the distribution. It is best suited for larger datasets

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(
    data=diamonds.sort_values("color"), x="color",
    y="price", kind="boxen"
)
```

#### Violinplots

Combine a boxplot with the kernel density estimation 

```python
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex",
    kind="violin"
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", split=True
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", split=True, inner="stick", palette="pastel"
)
```

## Bar plots

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(
    data=titanic, x="sex", y="survived", hue="class",
    kind="bar"
)
```

The default error bars show 95% confidence intervals

```python
sns.catplot(
    data=titanic, x="age", y="deck", kind="bar"
)
```

```python
sns.catplot(
    data=titanic, x="deck", kind="count",
    palette="ch:.25"
)
```

```python
sns.catplot(
    data=titanic, y="deck", hue="class", kind="count",
    palette="pastel", edgecolor=".6"
)
```

## Point plots

```python
sns.catplot(
    data=titanic, x="sex", y="survived",
    hue="class", kind="point"
)
```
