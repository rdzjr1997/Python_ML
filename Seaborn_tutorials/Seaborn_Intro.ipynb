## Seaborn Tutorial

```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# Create a visualization
sns.relplot(
    data=tips, x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size"
)
```

```python
# "kind" parameter of relplot
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False)
)
```

## Statistical estimation

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event"
)

# The error bars represent the uncertainty of the estimate.
# This is know as confidence intervals (confidence percentage?)
```

```python
# Linear regression model and its uncertainty
sns.lmplot(
    data=tips, x="total_bill", y="tip", col="time",
    hue="smoker"
)
```

## Distributional representations

Visualizing distributions of variables in the dataset.

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
# Plotting the empirical cumulative distribution function of the data
sns.displot(
    data=tips, kind="ecdf", x="total_bill",
    col="time", hue="smoker", rug=True
)
```

## Plots for categorical data

```python
# Swarm plot
sns.catplot(
    data=tips, kind="swarm", x="day",
    y="total_bill", hue="smoker"
)
```

```python
# KDE for distribution
sns.catplot(
    data=tips, kind="violin", x="day",
    y="total_bill", hue="smoker", split=True
)
```

```python
# Showing the mean value and its confidence interval
sns.catplot(
    data=tips, kind="bar", x="day",
    y="total_bill", hue="smoker"
)
```

## Multivariate views on complex datasets

Combination of multiple kinds of plots to give informative summaries of a dataset

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.jointplot(
    data=penguins, x="flipper_length_mm",
    y="bill_length_mm", hue="species"
)
```

```python
sns.pairplot(
    data=penguins, hue="species"
)
```
