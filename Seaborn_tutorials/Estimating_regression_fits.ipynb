```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

%matplotlib inline
```

## Drawing linear regression models

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.regplot(
    data=tips, x="total_bill", y="tip"
)
```

```python
sns.lmplot(
    data=tips, x="total_bill", y="tip"
)
```

#### Linear regression with one of the variables as discrete values

```python
sns.lmplot(
    data=tips, x="size", y="tip"
)
```

#### Plot of an estimate of central tendency along with a confidence interval

```python
sns.lmplot(
    data=tips, x="size", y="tip", x_estimator=np.mean
)
```

## Fitting different kinds of models

```python
anscombe = sns.load_dataset("anscombe")
anscombe.head()
```

```python
sns.lmplot(
    data=anscombe.query("dataset == 'I'"), x="x", y="y",
    ci=None, scatter_kws={"s": 80}
)
```

```python
sns.lmplot(
    data=anscombe.query("dataset =='II'"), x="x", y="y",
    ci=None, scatter_kws={"s": 80}
)
```

## Fitting a polynomial regression model for high-order relationships

```python
sns.lmplot(
    data=anscombe.query("dataset =='II'"), x="x", y="y",
    order=2, ci=None, scatter_kws={"s": 80}
)
```
