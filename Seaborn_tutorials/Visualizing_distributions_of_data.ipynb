```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

%matplotlib inline
```

## Ploting univariate histograms

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.displot(
    data=penguins, x="flipper_length_mm"
)
```

#### Choosing bin size

```python
sns.displot(
    data=penguins, x="flipper_length_mm",
    binwidth=3
)
```

#### Choosing number of the bins

```python
sns.displot(
    data=penguins, x="flipper_length_mm",
    bins=20
)
```

#### Awkward gaps in the distribution due the default bin width

```python
tips = sns.load_dataset("tips")
sns.displot(tips, x="size")
```

#### Array of bin breaks to bins

```python
sns.displot(
    data=tips, x="size",
    bins=[1, 2, 3, 4, 5, 6, 7]
)
```

#### Choosing bin breaks that represent the unique values by means of "discrete=True"

```python
sns.displot(
    data=tips, x="size", discrete=True
)
```

#### Visualizing categorical variable using the logic of a histogram

```python
sns.displot(
    data=tips, x="day", shrink=0.8
)
```

## Conditioning on other variables

#### Conditional subsetting via the "hue semantic"

```python
sns.displot(
    data=penguins, x="flipper_length_mm",
    hue="species"
)
```

#### Step plot

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    element="step"
)
```

#### Stacked histogram

```python
sns.displot(
    data=penguins, x="flipper_length_mm",
    hue="species", multiple="stack"
)
```

#### Moving the bars horizontally

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="sex",
    multiple="dodge"
)
```

```python
sns.displot(
    data=penguins, x="flipper_length_mm", col="sex"
)
```

## Normalized histogram statistics

#### Normalization applied to the entire distribution

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    stat="density"
)
```

#### Normalization applied to each subset independently

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    stat="density", common_norm=False
)
```

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    stat="probability"
)
```

## Kernel density estimation

#### A KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

```python
sns.displot(
    data=penguins, x="flipper_length_mm", kind="kde"
)
```

#### Choosing the smoothing bandwith

```python
sns.displot(
    data=penguins, x="flipper_length_mm", kind="kde",
    bw_adjust=0.25
)
```

```python
sns.displot(
    data=penguins, x="flipper_length_mm", kind="kde",
    bw_adjust=2
)
```

## Conditioning on other variables

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    kind="kde"
)
```

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    kind="kde", multiple="stack"
)
```

#### Changing opacity

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    kind="kde", fill=True
)
```

## Kernel density estimation pitfalls

#### KDE have problems with data that are naturally bounded

```python
sns.displot(
    data=tips, x="total_bill", kind="kde"
)
```

#### "cut" parameter -> specifies how far the curve should be extended beyond the extreme datapoints

```python
sns.displot(
    data=tips, x="total_bill", kind="kde", cut=0
)
```

```python
diamonds = sns.load_dataset("diamonds")
sns.displot(
    data=diamonds, x="carat", kind="kde"
)

```

```python
sns.displot(
    data=diamonds, x="carat"
)
```

#### Histograms and kde together

```python
sns.displot(
    data=diamonds, x="carat", kde=True
)
```

## Empirical cumulative distributions

Empirical cumulative distribution (ECDF)

Draws a monotonically-increasing curve through each datapoint such that the height of the curve reflects the proportion of observations with a smaller value



```python
sns.displot(
    data=penguins, x="flipper_length_mm", kind="ecdf"
)
```

```python
sns.displot(
    data=penguins, x="flipper_length_mm", hue="species",
    kind="ecdf"
)
```

## Visualizing bivariate distributions

Assigning a second variable to "y" will plot a bivariate distribution

```python
sns.displot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm"
)
```

```python
sns.displot(
    penguins, x="bill_length_mm", y="bill_depth_mm",
    kind="kde"
)
```

```python
sns.displot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    hue="species"
)
```

```python
sns.displot(
    penguins, x="bill_length_mm", y="bill_depth_mm",
    kind="kde", hue="species"
)
```

```python
sns.displot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    binwidth=(2, .5)
)
```

```python
sns.displot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    binwidth=(2, .5), cbar=True
)
```

## Distribution visualization in other settings

#### Plotting joint and marginal distributions

```python
sns.jointplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm"
)
```

```python
sns.jointplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    hue="species", kind="kde"
)
```

#### JointGrid class

```python
g = sns.JointGrid(
    data=penguins, x="bill_length_mm", y="bill_depth_mm"
)
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

#### "rug" plot to show marginal distributions

```python
sns.displot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    kind="kde", rug=True
)
```

#### Plotting many distributions

pairplot() -> Visualize the univariate distribution of all variables in a dataset along with all of their pairwise relationships

```python
sns.pairplot(data=penguins, hue="species")
```
